```python
import skimage.metrics
import torch
import torchvision.transforms as transforms
from PIL import Image
import os
import numpy as np


T = transforms.Compose([transforms.ToTensor()])
path = './dataset/OCTA_new/3mm_CROPPED'
path_2 = './dataset/OCTA_new/6mm_CROPPED'
filelist = os.listdir(path)
j = 1
Mean_y, Mean_gt, Var_y, Var_gt = [], [], [], []
for i in filelist:
    lr_name = os.path.join(path_2, str(j)+"_lr.png")
    hr_name = os.path.join(path, str(j)+"_hr.png")

    yimg = T(Image.open(lr_name).convert('L')).numpy()
    gtimg = T(Image.open(hr_name).convert('L')).numpy()
    Mean_y.append(np.mean(yimg))
    Mean_gt.append(np.mean(gtimg))
    Var_y.append(np.std(yimg))
    Var_gt.append(np.std(gtimg))

print(np.mean(np.array(Mean_y)), np.mean(np.array(Mean_gt)), np.mean(np.array(Var_y)), np.mean(np.array(Var_gt)))
```
